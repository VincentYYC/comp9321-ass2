```python
import pandas as pd
import numpy as np

df_rating = pd.read_csv('processdata/ratings.csv')
df_rating.head()
```

```python
df_movie = pd.read_csv('processdata/movies.csv')
df_movie.head()
```

```python
df = pd.merge(df_rating, df_movie, on='movieId')
df.head()
```

```python
movie_matrix = df.pivot_table(index = 'userId', columns = 'title', values = 'rating')
movie_matrix.head()
```

```python
AFO_user_rating = movie_matrix['Air Force One (1997)']
simliar_to_movie = movie_matrix.corrwith(AFO_user_rating)
simliar_to_movie
```

```python
corr_AFO = pd.DataFrame(simliar_to_movie, columns = ['Correlation'])
```

```python
corr_AFO.dropna(inplace = True) # remove NaN
corr_AFO
```

```python
ratings = pd.DataFrame(df.groupby('title')['rating'].mean())
ratings['number_of_ratings'] = df.groupby('title')['rating'].count()
```

```python
corr_AFO = corr_AFO.join(ratings['number_of_ratings'],how = 'left',lsuffix='_left', rsuffix='_right')
corr_AFO
```

```python
corr_AFO[corr_AFO['number_of_ratings']>100].sort_values(by = 'Correlation', ascending = False)
```
